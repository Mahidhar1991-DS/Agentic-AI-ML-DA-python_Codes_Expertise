```python
def convert_money(value):
    if isinstance(value, str):
        value = value.replace('€', '').replace(',', '')
        if 'M' in value:
            return float(value.replace('M', '')) * 1_000_000
        elif 'K' in value:
            return float(value.replace('K', '')) * 1_000
        else:
            return float(value)
    return value

money_cols = ['Value', 'Wage', 'Release Clause']

for col in money_cols:
    if col in fifa.columns:
        fifa[col] = fifa[col].apply(convert_money)
```

```python
numeric_cols = ['Age', 'Overall', 'Potential', 'Wage', 'Value', 'Release Clause']

numeric_cols = [col for col in numeric_cols if col in fifa.columns]

corr = fifa[numeric_cols].corr()

plt.figure(figsize=(7,5))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap (Numeric Columns)")
plt.show()
```

```python
cols = ['Age', 'Overall', 'Potential', 'Value', 'Wage']

plt.figure(figsize=(8,5))
sns.pairplot(fifa[cols], corner=True)
plt.show()
```

```python
cols = ['Age', 'Overall', 'Potential', 'Value', 'Wage']

corr = fifa[cols].corr()

sns.clustermap(corr, annot=True, cmap='coolwarm', figsize=(7, 7))
plt.title("Clustermap of Numeric Features")
plt.show()
```

```python
avg_overall_nat = fifa.groupby("Nationality")["Overall"].mean().sort_values(ascending=False)

print("Top 10 Nationalities by Average Overall Rating:\n")
print(avg_overall_nat.head(10))
```

```python
avg_overall_club = fifa.groupby("Club")["Overall"].mean().sort_values(ascending=False)

print("Top 10 Clubs by Average Overall Rating:\n")
print(avg_overall_club.head(10))
```

```python
avg_wage_club = fifa.groupby("Club")["Wage"].mean().sort_values(ascending=False)

print("Top 10 Clubs by Average Wage (€):\n")
print(avg_wage_club.head(10))
```

```python
fifa['Value'] = fifa['Value'].apply(convert_money)

# Group by Age
avg_value_age = fifa.groupby("Age")["Value"].mean().sort_index()

print("Average Player Value (€) by Age:\n")
print(avg_value_age)
```

```python
fifa['Value'] = fifa['Value'].apply(convert_money)
fifa['Wage'] = fifa['Wage'].apply(convert_money)

# Group calculations
foot_stats = fifa.groupby("Preferred Foot").agg(
    count=("Preferred Foot", "count"),
    avg_overall=("Overall", "mean"),
    avg_value=("Value", "mean"),
    avg_wage=("Wage", "mean")
).sort_values("avg_overall", ascending=False)

print("\nPreferred Foot Analysis:\n")
print(foot_stats)
```

```python
top_nat = fifa['Nationality'].value_counts().head(5).index
subset = fifa[fifa['Nationality'].isin(top_nat)]

# Multi-column grouping
multi_stats = subset.groupby(
    ['Nationality', 'Preferred Foot']
).agg(
    player_count=('Name', 'count'),
    avg_overall=('Overall', 'mean'),
    avg_value=('Value', 'mean'),
    avg_wage=('Wage', 'mean')
)

print("\nNationality + Preferred Foot Analysis:\n")
print(multi_stats)
```
